## Part 1

```python
import spacy

companies_name = companies_df['ticker'].unique()

model = spacy.load("en_core_web_sm")

def extract_company_names(text):
    ents = model(text)
    entities = [entity.text for entity in ents.ents if entity.text in companies_name]
    return entities

tweets_df['company_mentions'] = tweets_df['text'].apply(lambda x: extract_company_names(x))

company_mentions_flat = [mention for sublist in tweets_df['company_mentions'] for mention in sublist]

company_counts = pd.Series(company_mentions_flat).value_counts()
```

```python
top_companies = company_counts[:20]
low_companies = company_counts[-20:]

plt.figure(figsize=(10,6))
cmap = plt.get_cmap('hsv')
colors = cmap(np.linspace(0, 1, num=20))
sns.barplot(x=top_companies.index, y=top_companies.values, palette=colors)
plt.xticks(rotation=90)
plt.title('Top 20 cashtags')
plt.grid()
plt.show()

# Lower companies have 1 cashtags each!
# plt.figure(figsize=(10,6))
# cmap = plt.get_cmap('hsv')
# colors = cmap(np.linspace(0, 1, num=20))
# sns.barplot(x=top_companies.index, y=low_companies.values, palette=colors)
# plt.xticks(rotation=90)
# plt.title('Low 20 cashtags')
# plt.show()
```

## Part 2

```python
import matplotlib.dates as mdates

tweets_df['ticker'] = tweets_df['text'].str.findall(r'\$([A-Za-z]+)')
tweets_df['ticker'] = tweets_df['ticker'].apply(' '.join)
tweets_df['ticker'] = tweets_df['ticker'].apply(lambda x: x.upper())

selected_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'INTC']
tweets_df_sample = tweets_df[tweets_df['ticker'].isin(selected_tickers)]

tweets_df_sample['day'] = tweets_df_sample['created_at'].dt.to_period('D')
tweets_df_sample['day'] = tweets_df_sample['day'].dt.to_timestamp()

tweet_counts = tweets_df_sample.groupby(['ticker', 'day'])['id'].count().reset_index().rename(columns={'id': 'tweet_count'})

plt.figure(figsize=(15,5))
sns.lineplot(data=tweet_counts, x='day', y='tweet_count', hue='ticker')
plt.title('Tweet Count for Selected Stocks Over Time')
plt.xlabel('Date')
plt.ylabel('Tweet Count')
date_fmt = mdates.DateFormatter('%m-%d')
plt.gca().xaxis.set_major_formatter(date_fmt)
plt.grid()
plt.show()
```

```python
tweets_df['created_at'] = pd.to_datetime(tweets_df['created_at'])

tweets_df['hour'] = tweets_df['created_at'].dt.hour
tweets_per_hour = tweets_df.groupby('hour').size()

plt.figure(figsize=(10,6))
bars = plt.bar(tweets_per_hour.index, tweets_per_hour.values, color='green')

start_trading_hour = 10
end_trading_hour = 16
for i in range(start_trading_hour, end_trading_hour):
    bars[i].set_color('yellow')

plt.title('Number of tweets per hour (Yellow is trading hours)')
plt.xlabel('Hour of day')
plt.ylabel('Number of tweets')
plt.show()
```

```python
tweets_df['ticker'] = tweets_df['text'].str.findall(r'\$([A-Za-z]+)')
tweets_df['ticker'] = tweets_df['ticker'].apply(' '.join)
tweets_df['ticker'] = tweets_df['ticker'].apply(lambda x: x.upper())

selected_tickers = ['AAPL', 'TSLA']
tweets_df_sample = tweets_df[tweets_df['ticker'].isin(selected_tickers)]

tweets_df_sample['hour'] = tweets_df_sample['created_at'].dt.hour

retweets_df = tweets_df_sample[tweets_df_sample['retweeted_status_id'] != 0]

grouped = retweets_df.groupby(['ticker', 'hour']).size().reset_index(name='counts')

min_counts = pd.DataFrame()
min_counts['hour'] = np.arange(24)

for ticker in grouped['ticker'].unique():
    ticker_data = grouped[grouped['ticker'] == ticker]
    min_counts = pd.merge(min_counts, ticker_data, on='hour', how='left')

min_counts = min_counts.fillna(0).set_index('hour').min(axis=1)

plt.figure(figsize=(10,6))

tickers = grouped['ticker'].unique()
colors = ['cornflowerblue', 'lightcoral'] 

for i, ticker in enumerate(tickers):
    ticker_data = grouped[grouped['ticker'] == ticker]
    plt.bar(ticker_data['hour'], ticker_data['counts'], color=colors[i], label=ticker)
    plt.bar(ticker_data['hour'], min_counts, color='darkmagenta') 

plt.title('Number of retweets per hour for Apple and Tesla ( Khaki is trading hours)')
plt.xlabel('Hour of day')
plt.ylabel('Number of retweets')
plt.legend()
plt.axvspan(10, 15, facecolor='darkkhaki', alpha=0.3)
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tweets_df_sample['hour'] = tweets_df_sample['created_at'].dt.hour
tweets_df_sample_retweets = tweets_df_sample[tweets_df_sample['retweeted_status_id'] != 0]

grouped = tweets_df_sample_retweets.groupby(['hour', 'retweeted_status_id'])
retweets_per_tweet = grouped.size().reset_index(name='retweets')

grouped_all = tweets_df_sample_retweets.groupby(['hour', 'retweeted_status_id'])
retweets_per_tweet_all = grouped_all.size().reset_index(name='retweets')
avg_retweets_all = retweets_per_tweet_all.groupby('hour')['retweets'].sum() / retweets_per_tweet_all.groupby('hour')['retweeted_status_id'].nunique()

grouped_aapl = tweets_df_sample_retweets[tweets_df_sample_retweets['ticker'] == 'AAPL'].groupby(['hour', 'retweeted_status_id'])
retweets_per_tweet_aapl = grouped_aapl.size().reset_index(name='retweets')
avg_retweets_aapl = retweets_per_tweet_aapl.groupby('hour')['retweets'].sum() / retweets_per_tweet_aapl.groupby('hour')['retweeted_status_id'].nunique()

grouped_tsla = tweets_df_sample_retweets[tweets_df_sample_retweets['ticker'] == 'TSLA'].groupby(['hour', 'retweeted_status_id'])
retweets_per_tweet_tsla = grouped_tsla.size().reset_index(name='retweets')
avg_retweets_tsla = retweets_per_tweet_tsla.groupby('hour')['retweets'].sum() / retweets_per_tweet_tsla.groupby('hour')['retweeted_status_id'].nunique()

plt.plot(avg_retweets_all.index, avg_retweets_all.values, color='red', label='All tweets')
plt.plot(avg_retweets_aapl.index, avg_retweets_aapl.values, color='green', label='AAPL')
plt.plot(avg_retweets_tsla.index, avg_retweets_tsla.values, color='blue', label='TSLA')

plt.fill_between(range(10, 16), 1, 1.7, color='khaki', alpha=0.3)

plt.xlabel('Hour')
plt.ylabel('Retweets per Tweet')
plt.title('Retweets per tweet for AAPL and TSLA (Yellow is trading hours)')
plt.legend()
plt.grid()
plt.show()
```

## Part 6

```python
company_names = ['AMZN', 'NVDA', 'TWTR', 'LTC', 'NEO', 'SPY', 'DIA', 'MSFT', 'WMT', 'NFLX', 'BCC', 'OWCP', 'GE', 'IBM', 'TSLA', 'AAPL', 'AMD', 'IWM', 'INTC']

num = len(company_names)
covar = np.zeros((num, num))

for tweet in tweets_df.iterrows():
    for i in range(num):
        for j in range(num):
            if (i != j):
                if company_names[i] in tweet[1].ticker and company_names[j] in tweet[1].ticker:
                    covar[i,j] += 1
```

```python
upper_tri_indices = np.triu_indices(num, k=1)

upper_tri_flattened = covar[upper_tri_indices]
top_5_indices = np.argpartition(upper_tri_flattened, -5)[-5:]

top_5_pairs = np.array(upper_tri_indices)[:, top_5_indices]

labels = [f'{company_names[i]} & {company_names[j]}' for i, j in top_5_pairs.T]

values = upper_tri_flattened[top_5_indices]

palette = plt.get_cmap('Set2')

plt.figure(figsize=(10, 6))
plt.barh(labels, values, color=[palette(i) for i in range(5)])
plt.xlabel('Number of co-occurrences')
plt.title('Top 5 pairs of companies appearing together in tweets')
plt.grid()
plt.show()
```
