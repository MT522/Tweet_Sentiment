```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd

options = webdriver.EdgeOptions()
driver = webdriver.Edge(options=options)

driver.get("https://www.digikala.com/search/notebook-netbook-ultrabook/")

SCROLL_PAUSE_TIME = 2  
screen_height = driver.execute_script("return window.screen.height;") 
i = 1
while True:
    driver.execute_script(f"window.scrollTo(0, {screen_height * i});")
    i += 1
    time.sleep(SCROLL_PAUSE_TIME)

    scroll_height = driver.execute_script("return document.body.scrollHeight;")
    if screen_height * i > scroll_height:
        break

df = pd.DataFrame(columns=['title', 'price'])

soup = BeautifulSoup(driver.page_source, 'html.parser')

ultrabooks = soup.find_all('div', class_='product-list_ProductList__item__LiiNI')
for ultrabook in ultrabooks:
    title = ultrabook.find('h3', class_='styles_VerticalProductCard__productTitle__6zjjN').text.strip()
    price = ultrabook.find('span', attrs={'data-testid': 'price-final'}).text.strip()

    df = df.append({'title':title, 'price':price}, ignore_index=True)

driver.quit()
```

```python
df.to_csv('digikala_ultrabook.csv', encoding='utf-8-sig')
df
```

We have scraped the webpage thoroughly by scrolling down using a **selenium** web driver. The scrolled content is then passed to **Beautiful Soup** in order to extract the price based on the html structure of the webpage. You see that all data was successfully collected and saved in the dataframe named **"digikala_ultrabook.csv"**.
